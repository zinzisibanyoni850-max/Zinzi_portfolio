# Task
Perform exploratory data analysis, descriptive analysis, and prescriptive analysis with data visualization on the dataset provided in "/content/2020-2025.csv".

## Load data

### Subtask:
Load the dataset from "/content/2020-2025.csv" into a pandas DataFrame.


**Reasoning**:
Load the dataset into a pandas DataFrame and display the first few rows.



```python
import pandas as pd

df = pd.read_csv('/content/2020-2025.csv')
display(df.head())
```

## Exploratory data analysis (eda)

### Subtask:
Perform initial data exploration to understand the structure, content, and basic statistics of the dataset. This includes checking for missing values, data types, and unique values.


**Reasoning**:
Perform initial data exploration by checking data types, missing values, unique values, and descriptive statistics as per the instructions.



```python
df.info()
display(df.isnull().sum())
display(df.nunique())
display(df.describe())
```

## Descriptive analysis

### Subtask:
Calculate and present descriptive statistics for relevant columns (e.g., mean, median, standard deviation, quartiles).


**Reasoning**:
Calculate and display the mean, median, standard deviation, and quartiles for each of the year columns as requested in the instructions.



```python
year_columns = ['2020', '2021', '2022', '2023', '2024', '2025']

mean_values = df[year_columns].mean()
median_values = df[year_columns].median()
std_values = df[year_columns].std()
quartiles = df[year_columns].quantile([0.25, 0.5, 0.75])

descriptive_stats = pd.DataFrame({
    'Mean': mean_values,
    'Median': median_values,
    'Standard Deviation': std_values,
    '25th Percentile': quartiles.loc[0.25],
    '50th Percentile': quartiles.loc[0.50],
    '75th Percentile': quartiles.loc[0.75]
})

display(descriptive_stats)
```

## Data visualization

### Subtask:
Create visualizations to explore relationships between variables, identify patterns, and present key findings from the descriptive analysis.


**Reasoning**:
Create the requested line plots for mean, median, and standard deviation, and the box plot for distribution of values per year.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set_style("whitegrid")

# Create the line plot for mean values
plt.figure(figsize=(10, 6))
mean_values.plot(kind='line', marker='o')
plt.title('Trend of Mean Values Over Time')
plt.xlabel('Year')
plt.ylabel('Mean Value')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Create the line plot for median values
plt.figure(figsize=(10, 6))
median_values.plot(kind='line', marker='o', color='orange')
plt.title('Trend of Median Values Over Time')
plt.xlabel('Year')
plt.ylabel('Median Value')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Create the line plot for standard deviation values
plt.figure(figsize=(10, 6))
std_values.plot(kind='line', marker='o', color='green')
plt.title('Trend of Standard Deviation Over Time')
plt.xlabel('Year')
plt.ylabel('Standard Deviation')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Create the box plot for each year
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[year_columns])
plt.title('Distribution of Values per Year')
plt.xlabel('Year')
plt.ylabel('Value')
plt.tight_layout()
plt.show()
```
