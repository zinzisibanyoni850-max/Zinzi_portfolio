## Exploratory data analysis (eda)

### Subtask:
Display the first few rows and the data types of the columns. Check for missing values and handle them if necessary. Get a summary of the numerical and categorical features.


**Reasoning**:
Display the first few rows, data types, check for missing values, and display descriptive statistics for numerical and categorical columns.



```python
display(df.head())
display(df.info())
display(df.isnull().sum())
display(df.describe())
display(df.describe(include='object'))
```

**Reasoning**:
Based on the output of `df.info()` and `df.isnull().sum()`, the columns 'in_shazam_charts' and 'key' have missing values. The 'streams' and 'in_deezer_playlists' columns are of object type but should be numerical. The 'in_shazam_charts' column is also of object type and likely contains non-numeric values causing the missing values. The 'key' column is categorical with missing values. Before handling missing values, I need to convert the 'streams' and 'in_deezer_playlists' columns to numeric, coercing errors, and investigate the 'in_shazam_charts' column to see what non-numeric values are present.



```python
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')
df['in_deezer_playlists'] = pd.to_numeric(df['in_deezer_playlists'], errors='coerce')
display(df['in_shazam_charts'].value_counts(dropna=False))
```

**Reasoning**:
The 'in_shazam_charts' column contains some non-numeric values besides the missing ones, as indicated by the value counts. These non-numeric values and the missing values need to be handled by converting the column to numeric with coercion. The 'key' column has missing values and is categorical, so imputation with the most frequent value is a reasonable strategy. After handling these, I can re-run the descriptive statistics to confirm the changes and ensure the columns are now correctly typed for analysis.



```python
df['in_shazam_charts'] = pd.to_numeric(df['in_shazam_charts'], errors='coerce')
df['key'] = df['key'].fillna(df['key'].mode()[0])
display(df.info())
display(df.isnull().sum())
display(df.describe())
display(df.describe(include='object'))
```

**Reasoning**:
Based on the output, the 'streams', 'in_deezer_playlists', and 'in_shazam_charts' columns still contain missing values after attempting coercion to numeric. These missing values need to be handled appropriately. Since these columns represent counts or chart positions, imputing with the median might be a suitable strategy as it is less sensitive to outliers than the mean. After imputation, the missing values should be gone and the data types for these columns should be numeric, which can be confirmed by displaying the null counts and data types again.



```python
for col in ['streams', 'in_deezer_playlists', 'in_shazam_charts']:
    df[col] = df[col].fillna(df[col].median())

display(df.isnull().sum())
display(df.info())
display(df.describe())
display(df.describe(include='object'))
```

## Descriptive analysis

### Subtask:
Calculate descriptive statistics for numerical columns (mean, median, standard deviation, etc.). Analyze the distribution of categorical variables (counts, percentages).


**Reasoning**:
Calculate and display descriptive statistics for all numerical columns and iterate through each categorical column to display value counts.

## Data visualization

### Subtask:
Create visualizations to understand the distribution of key features (histograms, box plots). Visualize relationships between variables (scatter plots, correlation matrix). Create visualizations to support the descriptive analysis (bar plots for categorical data).


**Reasoning**:
Generate histograms for key numerical features to visualize their distributions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols_for_hist = ['streams', 'bpm', 'danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'liveness_%', 'speechiness_%']

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols_for_hist):
    plt.subplot(3, 3, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

**Reasoning**:
Create box plots for key numerical features to identify potential outliers and visualize the spread of data.



```python
numerical_cols_for_box = ['streams', 'bpm', 'danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'liveness_%', 'speechiness_%']

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols_for_box):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Box plot of {col}')
plt.tight_layout()
plt.show()
```

**Reasoning**:
Generate scatter plots to visualize the relationship between pairs of numerical variables.



```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(x='streams', y='in_spotify_playlists', data=df)
plt.title('Streams vs In Spotify Playlists')

plt.subplot(1, 2, 2)
sns.scatterplot(x='danceability_%', y='energy_%', data=df)
plt.title('Danceability % vs Energy %')

plt.tight_layout()
plt.show()
```

**Reasoning**:
Compute and visualize the correlation matrix for numerical columns using a heatmap to understand the linear relationships between all pairs of numerical features.



```python
numerical_df = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numerical_df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

**Reasoning**:
Create bar plots for key categorical columns to visualize the frequency distribution of their categories.



```python
categorical_cols_for_bar = ['released_year', 'key', 'mode']

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.countplot(y='released_year', data=df, order=df['released_year'].value_counts().index)
plt.title('Count of Songs by Released Year')

plt.subplot(1, 3, 2)
sns.countplot(y='key', data=df, order=df['key'].value_counts().index)
plt.title('Count of Songs by Key')

plt.subplot(1, 3, 3)
sns.countplot(y='mode', data=df, order=df['mode'].value_counts().index)
plt.title('Count of Songs by Mode')

plt.tight_layout()
plt.show()
```
