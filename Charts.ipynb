```python
import matplotlib.pyplot as plt
import pandas as pd

# Example dataset
data = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr", "May"],
    "Sales": [200, 250, 300, 280, 350],
    "Profit": [50, 70, 90, 85, 120]
})

# 1. Line Graph (changes over time)
plt.plot(data["Month"], data["Sales"], marker="o")
plt.title("Sales Over Time")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()

# 2. Bar Chart (comparisons)
plt.bar(data["Month"], data["Profit"])
plt.title("Profit by Month")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.show()

# 3. Pie Chart (proportions)
plt.pie(data["Profit"], labels=data["Month"], autopct="%1.1f%%")
plt.title("Profit Share by Month")
plt.show()

# 4. Scatter Plot (relationships)
plt.scatter(data["Sales"], data["Profit"])
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()

# 5. Histogram (distribution)
ages = [22,25,26,28,30,32,35,36,40,41,42,45,50]
plt.hist(ages, bins=6)
plt.title("Age Distribution")
plt.xlabel("Age Range")
plt.ylabel("Frequency")
plt.show()
```
