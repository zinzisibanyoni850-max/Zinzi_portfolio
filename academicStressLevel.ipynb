## Data visualization (eda)

### Subtask:
Create visualizations to understand the data distribution and relationships between variables.


**Reasoning**:
Create count plots for the specified categorical columns to visualize their distributions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

categorical_cols = ['Your Academic Stage', 'Study Environment', 'What coping strategy you use as a student?', 'Do you have any bad habits like smoking, drinking on a daily basis?']

for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, y=col, palette='viridis')
    plt.title(f'Distribution of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```

**Reasoning**:
Create box plots to visualize the relationship between 'Your Academic Stage' and the specified numerical columns.



```python
numerical_cols = ['Peer pressure', 'Academic pressure from your home', 'What would you rate the academic  competition in your student life', 'Rate your academic stress index ']

for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x='Your Academic Stage', y=col, palette='viridis')
    plt.title(f'Relationship between Your Academic Stage and {col}')
    plt.xlabel('Your Academic Stage')
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```

**Reasoning**:
Create box plots to visualize the relationship between 'Study Environment' and the specified numerical columns.



```python
numerical_cols = ['Peer pressure', 'Academic pressure from your home', 'What would you rate the academic  competition in your student life', 'Rate your academic stress index ']

for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x='Study Environment', y=col, palette='viridis')
    plt.title(f'Relationship between Study Environment and {col}')
    plt.xlabel('Study Environment')
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```

**Reasoning**:
Create a correlation heatmap of the numerical columns to visualize the relationships between them.



```python
numerical_cols = ['Peer pressure', 'Academic pressure from your home', 'What would you rate the academic  competition in your student life', 'Rate your academic stress index ']
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Columns')
plt.tight_layout()
plt.show()
```

## Descriptive analysis

### Subtask:
Describe the characteristics of the data, such as central tendency, dispersion, and shape of the distribution.


**Reasoning**:
Calculate and display descriptive statistics for the numerical columns and value counts for the categorical columns to characterize the data.



```python
numerical_cols = ['Peer pressure', 'Academic pressure from your home', 'What would you rate the academic  competition in your student life', 'Rate your academic stress index ']
print("Descriptive Statistics for Numerical Columns:")
display(df[numerical_cols].agg(['mean', 'median', 'std']))

categorical_cols = ['Your Academic Stage', 'Study Environment', 'What coping strategy you use as a student?', 'Do you have any bad habits like smoking, drinking on a daily basis?']
print("\nValue Counts for Categorical Columns:")
for col in categorical_cols:
    print(f"\nValue Counts for '{col}':")
    display(df[col].value_counts())

print("\nInterpretation of Descriptive Statistics and Value Counts:")
print("The mean, median, and standard deviation provide insights into the central tendency, spread, and variability of the numerical variables.")
print("The value counts show the frequency distribution of categories within the categorical variables, highlighting the most common and least common responses.")
```
