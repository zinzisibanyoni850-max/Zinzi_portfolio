# Task
Perform exploratory data analysis, descriptive analysis, and prescriptive analysis with data visualization on the dataset provided in "/content/crocodile_dataset.csv".

## Load the data

### Subtask:
Load the `crocodile_dataset.csv` file into a pandas DataFrame.


**Reasoning**:
Import the pandas library and load the dataset into a DataFrame.



```python
import pandas as pd

df = pd.read_csv('/content/crocodile_dataset.csv')
```

## Exploratory data analysis (eda)

### Subtask:
Perform initial exploratory data analysis including displaying the first few rows, checking data types and non-null values, generating descriptive statistics, checking for missing values, and visualizing feature distributions.


**Reasoning**:
Display the first 5 rows, check data types and non-null values, generate descriptive statistics, check for missing values, and create histograms and box plots for numerical columns to perform initial exploratory data analysis.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Display the first 5 rows
print("First 5 rows of the DataFrame:")
display(df.head())

# 2. Print information about the DataFrame
print("\nDataFrame Info:")
df.info()

# 3. Generate descriptive statistics for numerical columns
print("\nDescriptive Statistics for Numerical Columns:")
display(df.describe())

# 4. Check for missing values
print("\nMissing values per column:")
display(df.isnull().sum())

# Select numerical columns for visualization
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# 5. Create histograms for numerical columns
print("\nHistograms of Numerical Columns:")
if len(numerical_cols) > 0:
    df[numerical_cols].hist(figsize=(15, 10))
    plt.tight_layout()
    plt.show()
else:
    print("No numerical columns found for histograms.")


# 6. Create box plots for numerical columns
print("\nBox Plots of Numerical Columns:")
if len(numerical_cols) > 0:
    plt.figure(figsize=(15, 10))
    for i, col in enumerate(numerical_cols):
        plt.subplot(len(numerical_cols), 1, i + 1)
        sns.boxplot(x=df[col])
        plt.title(f'Box Plot of {col}')
    plt.tight_layout()
    plt.show()
else:
    print("No numerical columns found for box plots.")
```

## Descriptive analysis

### Subtask:
Perform descriptive analysis including calculating summary statistics for relevant columns, analyzing relationships between variables using correlation matrices and scatter plots, and grouping data and calculating statistics based on categories if applicable.


**Reasoning**:
Calculate and display summary statistics for the numerical columns, calculate and display the correlation matrix, create a scatter plot, and calculate and display grouped summary statistics as per the subtask instructions.



```python
# 1. Calculate and display summary statistics for numerical columns
print("Summary Statistics for Numerical Columns:")
numerical_cols = ['Observed Length (m)', 'Observed Weight (kg)']
display(df[numerical_cols].agg(['mean', 'median', 'std', 'var', 'min', 'max']).T)
display(df[numerical_cols].quantile([0.25, 0.5, 0.75]).T.rename(columns={0.25: '25th Percentile', 0.5: 'Median (50th Percentile)', 0.75: '75th Percentile'}))
display(df[numerical_cols].mode().rename(index={0: 'Mode'}))


# 2. Calculate and display the correlation matrix for the numerical columns
print("\nCorrelation Matrix for Numerical Columns:")
correlation_matrix = df[numerical_cols].corr()
display(correlation_matrix)

# 3. Create a scatter plot to visualize the relationship between Observed Length (m) and Observed Weight (kg)
print("\nScatter Plot of Observed Length vs. Observed Weight:")
plt.figure(figsize=(10, 6))
plt.scatter(df['Observed Length (m)'], df['Observed Weight (kg)'])
plt.title('Scatter Plot of Observed Length vs. Observed Weight')
plt.xlabel('Observed Length (m)')
plt.ylabel('Observed Weight (kg)')
plt.grid(True)
plt.show()

# 4. Calculate and display summary statistics grouped by Conservation Status and Habitat Type
print("\nSummary Statistics Grouped by Conservation Status:")
display(df.groupby('Conservation Status')[numerical_cols].agg(['count', 'mean', 'median']))

print("\nSummary Statistics Grouped by Habitat Type:")
display(df.groupby('Habitat Type')[numerical_cols].agg(['count', 'mean', 'median']))
```

## Prescriptive analysis

### Subtask:
Identify potential areas for improvement or action based on the descriptive analysis findings and suggest potential interventions or strategies based on the data.


**Reasoning**:
Identify notable trends or patterns from the descriptive analysis findings related to observed length and weight across different conservation statuses and habitat types.



```python
print("Average Observed Length and Weight by Conservation Status:")
display(df.groupby('Conservation Status')[['Observed Length (m)', 'Observed Weight (kg)']].mean())

print("\nAverage Observed Length and Weight by Habitat Type:")
display(df.groupby('Habitat Type')[['Observed Length (m)', 'Observed Weight (kg)']].mean())
```

## Data visualization

### Subtask:
Create visualizations to support the findings from EDA and descriptive analysis (e.g., bar charts, line plots, scatter plots, heatmaps).


**Reasoning**:
Generate bar charts for average length, average weight, and observation counts grouped by Conservation Status and Habitat Type, and also generate a heatmap for the correlation matrix.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Average observed length by 'Conservation Status'
plt.figure(figsize=(12, 6))
sns.barplot(x='Conservation Status', y='Observed Length (m)', data=df, estimator='mean', ci=None)
plt.title('Average Observed Length by Conservation Status')
plt.xlabel('Conservation Status')
plt.ylabel('Average Observed Length (m)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 2. Average observed weight by 'Conservation Status'
plt.figure(figsize=(12, 6))
sns.barplot(x='Conservation Status', y='Observed Weight (kg)', data=df, estimator='mean', ci=None)
plt.title('Average Observed Weight by Conservation Status')
plt.xlabel('Conservation Status')
plt.ylabel('Average Observed Weight (kg)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 3. Count of observations by 'Conservation Status'
plt.figure(figsize=(12, 6))
sns.countplot(x='Conservation Status', data=df)
plt.title('Count of Observations by Conservation Status')
plt.xlabel('Conservation Status')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 4. Average observed length by 'Habitat Type'
plt.figure(figsize=(14, 7))
sns.barplot(x='Habitat Type', y='Observed Length (m)', data=df, estimator='mean', ci=None)
plt.title('Average Observed Length by Habitat Type')
plt.xlabel('Habitat Type')
plt.ylabel('Average Observed Length (m)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 5. Average observed weight by 'Habitat Type'
plt.figure(figsize=(14, 7))
sns.barplot(x='Habitat Type', y='Observed Weight (kg)', data=df, estimator='mean', ci=None)
plt.title('Average Observed Weight by Habitat Type')
plt.xlabel('Habitat Type')
plt.ylabel('Average Observed Weight (kg)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 6. Count of observations by 'Habitat Type'
plt.figure(figsize=(14, 7))
sns.countplot(x='Habitat Type', data=df)
plt.title('Count of Observations by Habitat Type')
plt.xlabel('Habitat Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 7. Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Observed Length and Weight')
plt.show()
```
